```python
#Scatter plot
plt.figure(figsize=(15,15))
covid.plot.scatter(x="total_cases_per_million", y="total_deaths_per_million", s=10)
```

```python
#cleared scatter plot
covid.plot.scatter(x="total_cases_per_million", y="total_deaths_per_million", s=20, color="red", fontsize=15, marker="*")
plt.rcParams['figure.figsize']=(10,5)
plt.xlabel('total_cases_per_million',fontsize=15)
plt.ylabel('total_deaths_per_million',fontsize=15)
plt.title("Total cases per million vs Total deaths per million", fontsize=15)
plt.xlim(10000,8000)
plt.ylim(0,1200)
plt.grid()
```

```python
#subplot between total_deaths_per_million and total_cases_per_million
plt.figure(figsize=(13,5))
plt.subplot(1,2,1)
plt.plot(covid["total_deaths_per_million"],color="firebrick")
plt.title("Total deaths/million",fontsize=14)
plt.subplot(1,2,2)
plt.plot(covid["total_cases_per_million"],color="coral")
plt.title("Total cases/million",fontsize=14)
plt.suptitle("Total deaths/million & Total cases/million", fontsize=17)
```

```python
#Subplot between new_tests_per_thousand and tests_units
plt.figure(figsize=(17,5))
plt.subplot(1,2,1)
plt.plot(covid["new_tests_per_thousand"],color="maroon")
plt.title("New tests per thousand",fontsize=14)
plt.subplot(1,2,2)
plt.hist(covid["tests_units"],color="mediumspringgreen",width=0.5)
plt.xticks(rotation=30)
plt.title("tests_units",fontsize=14)
plt.suptitle("Total deaths/million & Life Expectancy", fontsize=17)
```

```python
#linear plot
covid.groupby('continent').sum().plot(fontsize=10)
plt.rcParams['figure.figsize']=(10,10)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=1)
plt.ylim(0,10000)
plt.grid()
```

```python
#horizontal plot
plt.figure(figsize=(10,7))
covid["total_tests"].describe().plot(kind="barh", label="total_tests",color="red",fontsize=15)
covid["new_tests"].describe().plot(kind="barh",label="new_tests",color="orange")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=1)
plt.xticks(rotation=30)
plt.xlim(0.0,400000)
```
